Let's move further and have an analysis of box plot and histogram combinations for few columns

```python
# for this, we'll create a function to plot both of them together

def histogram_boxplot(data, feature, figsize=(8,3), kde = True, bins = 10):
    """
    Boxplot and histogram combined

    data: dataset
    feature: dataset column
    figsize: size of figure (defaulted to (8,3))
    kde: whether to show the density curve (default False)
    bins: number of bins for histogram (defaulted to 10)
    """
    f2, (ax_box2, ax_hist2) = plt.subplots(
        nrows = 2,  # Number of rows of the subplot grid = 2
        sharex = True,  # x-axis will be shared among all subplots
        gridspec_kw = {"height_ratios": (0.25, 0.75)},
        figsize = figsize)  # creating the 2 subplots
    
    sns.boxplot(data = data, x = feature, ax = ax_box2, showmeans = True, color = "green")  # boxplot will be created and a dot will indicate the mean value of the column
    
    if bins:
      sns.histplot(data = data, x = feature, kde = kde, ax = ax_hist2, bins = bins, color = "pink")
    else: 
      sns.histplot(data = data, x = feature, kde = kde, ax = ax_hist2, color = "pink")  # For histogram
    
    ax_hist2.axvline(data[feature].mean(), color = "purple", linestyle = "--")  # Add mean to the histogram
    
    ax_hist2.axvline(data[feature].median(), color = "blue", linestyle = "-")  # Add median to the histogram
```

> **SOLUTION (4 A.)**

```python
# To check if there is any significant difference between Funds raised by companies that are still operating vs companies
# that closed down, let's split the data
operating_companies = company[company['OperatingState'] == 'Operating']['Funds_in_million']
closed_companies = company[company['OperatingState'] == 'Closed']['Funds_in_million']

# Performing two-sample t-test as we want to compare the means of these samples to determine the ask
t_statistic, p_value = stats.ttest_ind(operating_companies, closed_companies)

# Print the results
print("t-statistic:", format(t_statistic, ".2f"))
print("p-value:", format(p_value, ".2f"))

# Check significance at alpha = 0.05, i.e., taking level of significance as 5%
alpha = 0.05

if p_value < alpha:
    print("There is a significant difference between funds raised by companies that are still operating and companies that closed down.")
else:
    print("There is no significant difference between funds raised by companies that are still operating and companies that closed down.")
```

> **SOLUTION (4 B.)**

**Null Hypothesis (Ho) -** There is no significant difference in the mean funds raised by companies that are still operating and companies that closed down.

**Alternative hypothesis (Ha) -** There is a significant difference in the mean funds raised by companies that are still operating and companies that closed down.

This is based on assumption that the mean value for funds raised by the two groups (operating companies and closed companies) is equal. The above conducted t-test is a perfect example for the Hypothesis analysis and there the Null Hypothesis is true.

> **SOLUTION (4 C.)**

Talking about **Test for Significance**:
 1. it includes formulating the Null/Alternate Hypothesis which we have already done above
 2. it checks for the level of significance which we can choose based on our assumption of how much chance of error we are willing to accept. Here I chose 5% chances which concluded the value of α = 0.05
 3. we can then determine the t-statistic whcih can be calculated from the data we have


**Conclusion -**
<br>The conclusion which can be drawn is that since the p-value is more than α, we fail to reject the null hypothesis and the observed difference in the mean funds can be due to random variance in the data however, there is no significant difference in the mean funds raised by companies that are still operating and companies that closed down.

> **SOLUTION (4 G.)**

**Null Hypothesis (Ho) -** The proportion of companies that are operating among winners is the same as the proportion of companies that are operating among contestants

**Alternate Hypothesis (Ha) -** The proportion of companies that are operating among winners is different from the proportion of companies that are operating among contestants

> **SOLUTION (4 H.)**

```python
# Since here we are comparing the 2 proportions, Chi-square test would be appropriate
from scipy.stats import chi2_contingency

# Create a contingency table
contingency_table = pd.crosstab(index=company_copy['OperatingState'], columns=company_copy['Result'])
contingency_table
```

```python
# getting the chi square statistic and p value

chi_2_stats, p_value, dof, exp_freq = chi2_contingency(contingency_table)

print("Chi-Square Statistic:", chi_2_stats)
print("P-value:", p_value)
print("Degree of Freedom:", dof)
print("Expected Frequency:", exp_freq)
```

```python
alpha = 0.05 # taking the level of significance to be 5%

if p_value < alpha:
    print("Rejecting the null hypothesis!")
else:
    print("Failed to reject the null hypothesis!")
```

**Conclusion:**

The proportion of companies that are operating among winners is different from the proportion of companies that are operating among contestants
